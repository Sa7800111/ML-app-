```python
train  = pd.read_csv(DATA / "train.csv")
test   = pd.read_csv(DATA / "test.csv")
gender = pd.read_csv(DATA / "gender_submission.csv")
```

```python
def missing_report(df):
    miss_cnt = df.isna().sum()
    miss_pct = df.isna().mean() * 100
    return pd.DataFrame({"missing_count": miss_cnt, "missing_percent": miss_pct}).sort_values("missing_percent", ascending=False)
```

```python
print("Missing Data - Train")
print(missing_report(train))
print("\nMissing Data - Test")
print(missing_report(test))
```
